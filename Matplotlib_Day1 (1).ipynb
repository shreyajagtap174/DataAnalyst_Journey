#Matplotlib

import matplotlib.pyplot as plt

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(1)
```

# Line Plot

x - numerical
y -  numerical

```python
plt.figure(figsize = (6,5))   # its figure size

plt.plot(titanic['age'], titanic['fare'] , color = 'g',
         linestyle = 'dotted' , marker = 'o' , linewidth = 2)

plt.xlabel('Age of passanger')
plt.ylabel('Fare of passanger')
plt.title('Fair paid by passanger with their age')
```

# BarPlot

x - catagorical
y - numerical(count,frequency)

```python
titanic['sex'].unique()
```

```python
gender_counts = titanic['sex'].value_counts()
```

```python
gender_counts
```

```python
x = gender_counts.index
```

```python
y = gender_counts.values
```

```python
plt.figure(figsize = (6,6))

plt.bar(x,y, color = ['blue', 'pink'], edgecolor = 'black')

plt.xlabel("Gender of passanger")
plt.ylabel("Gender of passanger")
plt.title("number of Male and Female Passanger")
```

```python
plt.figure(figsize=(8,5))

plt.hist(titanic['age'] , edgecolor = 'black' , bins = 40)

plt.xlabel("Age of passanger")
plt.ylabel("Count of passanger")
plt.title("Age Distribution of passanger")
```

# Box Plot

numerical

```python
plt.figure(figsize=(6,5))

plt.boxplot(titanic['fare'], vert=False)
plt.xlabel("Fare of passanger")
plt.title("Box plot of fare price")
```
