#Matplotlib.

import matplotlib.pyplot as plt

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(1)
```

# Line Plot

x - numerical
y - numerical

```python
plt.figure(figsize= (6, 5))  # its figure size

plt.plot(titanic['age'], titanic['fare'] , color = 'g',
         linestyle = 'dotted' , marker = 'o', linewidth= 2 )
plt.xlabel('Age of passenger')
plt.ylabel('Fare of passenger')
plt.title('Fare paid by passenger with their Age')
```

#Barplot
 x - categorical

 y - numerical (count, frequency )

```python
titanic['sex'].unique()
```

```python
gender_counts = titanic['sex'].value_counts()
```

```python
gender_counts
```

```python
x = gender_counts.index
```

```python
y = gender_counts.values
```

```python
plt.figure(figsize= (6,6))

plt.bar(x, y, color =['blue', 'pink'], edgecolor = 'black')

plt.xlabel("Gender of passenger")
plt.ylabel("Count of passenger")
plt.title("Number of Male and Female Passenger")
```

#Histogram -
 numerical - numerical - bins

```python
plt.figure(figsize= (8,5))

plt.hist(titanic['age'] , bins = 40, edgecolor = 'black', color ='g')

plt.xlabel("Age of passenger")
plt.ylabel("Count of passenger")
plt.title("Age Distribution of Passenger")
```

#BOX PLOT

Numerical

```python
plt.figure(figsize=(6,5))

plt.boxplot(titanic['fare'], vert=False)
plt.xlabel("Fare of passenger")
plt.title("Box plot of fare price")
```
